```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#turn grades into numeric for proper sorting later on
school_data_complete["grade"] = school_data_complete["grade"].replace({"9th":"9","10th":"10","11th":"11","12th":"12"}).astype(int)
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Printing summary and making a new dataframe
#Total Schools
schools = school_data_complete["school_name"].unique()
total_schools = len(schools)
print("Total Schools: "+str(total_schools))
#Total Students
students = school_data_complete["student_name"].unique()
total_students = len(students)
print("Total Students: "+str(total_students))
#Total Budget
budgets = school_data_complete["budget"].unique()
total_budgets = sum(budgets)
print("Total Budget: "+str(total_budgets))
#Average Math Score:
math_scores = school_data_complete["math_score"]
avg_math = math_scores.mean()
print("Average Math Score: "+"{:.2f}".format(avg_math))
#Average Reading Score
reading_scores = school_data_complete["reading_score"]
avg_reading = reading_scores.mean()
print("Average Reading Score: "+"{:.2f}".format(avg_reading))
#Overall Passing Rate
passing_rate = (reading_scores.mean()+math_scores.mean())/2
print("Overall Passing Rate: "+"{:.2f}".format(passing_rate))
#Percent Passing Math
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70,:]
math_pass_percent = 100*len(passing_math)/len(math_scores)
print("Percent of Passing Math Scores: "+"{:.2%}".format(math_pass_percent))
# Percent Passing Reading
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70,:]
reading_pass_percent = 100*len(passing_reading)/len(reading_scores)
print("Percent of Passing Reading Scores: "+"{:.2%}".format(reading_pass_percent))
#Place it all in a data frame.
school_summary = pd.DataFrame({"Total_Schools":total_schools,"Total_Students":total_students,
                              "Total_Budgets":total_budgets,"Average_Math_Score":avg_math,
                              "Average_Reading_Score":avg_reading,"Overall_Passing_Rate":passing_rate,
                              "Percent_Passing_Math":math_pass_percent,"Percent_Passing_Reading":reading_pass_percent},[0])
#Format Columns
school_summary["Total_Budgets"] = school_summary["Total_Budgets"].map("${:,}".format)
school_summary["Average_Math_Score"] = school_summary["Average_Math_Score"].map("{:.2f}".format)
school_summary["Average_Reading_Score"] = school_summary["Average_Reading_Score"].map("{:.2f}".format)
school_summary["Overall_Passing_Rate"] = school_summary["Overall_Passing_Rate"].map("{:.2f}".format)
school_summary["Percent_Passing_Math"] = school_summary["Percent_Passing_Math"].map("{:.2f}%".format)
school_summary["Percent_Passing_Reading"] = school_summary["Percent_Passing_Reading"].map("{:.2f}%".format)
school_summary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Take original dataframe and group by school name
school_data_grouped = school_data_complete.groupby("school_name")
# school_data_grouped["reading_score"]
```

```python
#Take in only the test scores to find the passing scores
subset_grades = school_data_complete[["school_name","math_score","reading_score"]]
#Use loc to find only those rows that passed.
pass_math = (subset_grades.loc[subset_grades["math_score"]>69])
pass_reading = (subset_grades.loc[subset_grades["reading_score"]>69])
#This grouping makes it so you can use count() for the calculation in the next section.
pass_math_grouped = pass_math.groupby("school_name")
pass_reading_grouped = pass_reading.groupby("school_name")
```

```python
#Use agg to fill in a new dataframe
school_data_agg = school_data_grouped.agg({"student_name":'count',"budget":'max',"math_score":'mean',
                                           "reading_score":'mean',"type":"max"})
#For some reason, mean did not work in the agg function so I needed to do this separately
school_data_agg["budget_per_student"]= school_data_agg["budget"]/school_data_agg["student_name"]
#Use the groupBy's we made before to calculate percent passing
school_data_agg["pass_math_percent"] = pass_math_grouped["math_score"].count()/school_data_grouped["math_score"].count()

school_data_agg["pass_reading_percent"] = pass_reading_grouped["reading_score"].count()/school_data_grouped["reading_score"].count()

school_data_agg["overall_pass_rate"]= (school_data_agg["pass_reading_percent"]+school_data_agg["pass_math_percent"])/2

#Rename Columns
school_data_agg = school_data_agg.rename(columns={"student_name":"student_count","budget":"total_budget","math_score":"avg_math_score",
                                                  "reading_score":"avg_reading_score","type":"school_type"})

school_data_agg.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#trim the school_data_agg so it only includes what we need
school_data_spending = school_data_agg[["avg_math_score","avg_reading_score","pass_math_percent",
                                       "pass_reading_percent","overall_pass_rate","budget_per_student"]]
#reset index so we can cut and groupBy by spending
school_data_spending = school_data_spending.reset_index()
school_data_spending["school_spending"] = pd.cut(school_data_spending["budget_per_student"],spending_bins,labels=group_names)
school_data_spending = school_data_spending.groupby("school_spending").mean()
del school_data_spending["budget_per_student"]
school_data_spending.head()
```
